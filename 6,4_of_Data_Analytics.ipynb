```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Read the CSV file
df = pd.read_csv('real_estate.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Drop rows with missing values
df = df.dropna()
```

```python
print(df.head())
```

```python
import numpy as np  # Import NumPy and alias it as np
```

```python
# Extract the independent and dependent variables using the correct column names
X = df[['X3 distance to the nearest MRT station']].values
y = df['Y house price of unit area'].values

# Create a scatter plot
sns.set_theme()
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.flatten(), y=y)
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Scatter Plot of Distance to MRT vs House Price per Unit Area')
plt.show()

# Perform linear regression
model = LinearRegression()
model.fit(X, y)

# Get the regression line
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)  # Correctly using np.linspace
y_range = model.predict(x_range)

# Plot the scatter plot with the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.flatten(), y=y)
plt.plot(x_range, y_range, color='red')
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Scatter Plot with Linear Regression Line')
plt.show()
```

```python
# Predict y values using the linear regression model
y_pred = model.predict(X)
```

```python
# Create a dataframe to compare real and predicted values
comparison_df = pd.DataFrame({'Real Values (y)': y, 'Predicted Values (y_pred)': y_pred})  # Use pd.DataFrame() from pandas
print("\nComparison of Real and Predicted Values:")
print(comparison_df.head())
```
